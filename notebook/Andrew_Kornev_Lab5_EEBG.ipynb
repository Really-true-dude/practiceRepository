```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
## Load the dataset
lab2_data = sns.load_dataset("titanic")

lab2_data.head()
```

```python
## Plot surviving dependency based on passegers class
sns.barplot(data=lab2_data, x="class", y="survived", hue="class")
plt.show()
```

```python
## Adding information about the people origin
sns.barplot(data=lab2_data, x="class", y="survived", hue="who")
plt.show()
```

```python
## Adding information about children's sex in the dataframe
lab2_data.loc[(lab2_data["who"] == "child") & (lab2_data["sex"] == "male"), "who"] = "boy"
lab2_data.loc[(lab2_data["who"] == "child") & (lab2_data["sex"] == "female"), "who"] = "girl"

# Plotting the results
sns.barplot(data=lab2_data, x="class", y="survived", hue="who")
plt.show()
```

```python
## Set confidence interval to 90% and length of the error bars
# Using errorbar as ci parameter is deprecated
sns.barplot(data=lab2_data, x="class", y="survived", hue="who", errorbar=("ci", 90), capsize=0.05)
plt.show()
```

```python
## Plotting the age of who survived and who not
sns.barplot(data=lab2_data, x="survived", y="age", hue="who", errorbar=("ci", 90), capsize=0.05)
plt.show()
```

```python
## Adding column to dataframe (survived_str)
lab2_data.loc[(lab2_data["survived"] == 1), "survived_str"] = "yes"
lab2_data.loc[(lab2_data["survived"] == 0), "survived_str"] = "no"


## Plotting results on new column
sns.barplot(data=lab2_data, x="survived_str", y="age", hue="who")
```

```python
## Change errorbars - (represent sd, length)
# Changing sample mean to median
import numpy as np
sns.barplot(data=lab2_data, x="survived_str", y="age", hue="who", errorbar=("ci", 90), capsize=0.05, estimator=np.median)
plt.show()
```

```python
## Boxplot
sns.boxplot(data=lab2_data, x="survived_str", y="age", hue="who")
plt.show()
```

```python
## Stripplot
sns.stripplot(data=lab2_data, x="survived_str", y="age", hue="who", jitter=0.3, alpha=0.6, dodge="true")
plt.show()
```
